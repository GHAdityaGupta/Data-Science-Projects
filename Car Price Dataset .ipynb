### Model Building - Feature Selection

```python
X = car1[num_col].drop('price',axis=1)
y = car1['price']
```

#### Recursive Feature Elimination with Random Forest Regressor

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
car1['fuelsystem']=le.fit_transform(car1['fuelsystem'])
car1['cylindernumber']=le.fit_transform(car1['cylindernumber'])
car1['enginetype']=le.fit_transform(car1['enginetype'])
car1['enginelocation']=le.fit_transform(car1['enginelocation'])
car1['drivewheel']=le.fit_transform(car1['drivewheel'])
car1['carbody']=le.fit_transform(car1['carbody'])
car1['doornumber']=le.fit_transform(car1['doornumber'])
car1['aspiration']=le.fit_transform(car1['aspiration'])
car1['fueltype']=le.fit_transform(car1['fueltype'])
car1['CarName']=le.fit_transform(car1['CarName'])
```

```python
X = car1.drop(columns = ['CarName','price'], axis=1)
y = car1['price']
```

```python
# Create the RFE object and rank each pixel
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
clf_rf_3 = RandomForestRegressor()      
rfe = RFE(estimator=clf_rf_3, n_features_to_select=15, step=1)
rfe = rfe.fit(X, y)
print('Chosen best 15 feature by rfe:',X.columns[rfe.support_])
```

```python
features=list(X.columns[rfe.support_])
```

### DataModelling and Evaluation

```python
x = X[features]
y = car1.price
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.30, random_state = 0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(x_train,y_train)
y_train_pred = lreg.predict(x_train)
y_test_pred = lreg.predict(x_test)
lreg.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor(random_state=0)
dt_regressor.fit(x_train,y_train)
y_train_pred = dt_regressor.predict(x_train)
y_test_pred = dt_regressor.predict(x_test)
dt_regressor.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
Rf = RandomForestRegressor(n_estimators = 15,
                              criterion = 'mse',
                              random_state = 20,
                              n_jobs = -1)
Rf.fit(x_train,y_train)
Rf_train_pred = Rf.predict(x_train)
Rf_test_pred = Rf.predict(x_test)
from sklearn.metrics import r2_score

r2_score(y_test,Rf_test_pred)
```
